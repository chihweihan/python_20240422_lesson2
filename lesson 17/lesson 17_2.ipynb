```python
import numpy as np
import pandas as pd

#後續都用dataframe清理資料
df1= pd.read_csv('world.csv')
df1
```

```python
df1.info()
```

```python
# 選取欄列 -> reindex
df2 = df1.reindex(columns=['洲名','國家','日期','總確診數','新增確診數'])
df2.info()
```

```python
#檢查 query(全是字串)

taiwan_df = df2.query('國家=="台灣"')
taiwan_df
```

```python
taiwan_df.query('新增確診數==0')
```

```python
tw2021_df = taiwan_df.query('日期>="2021-01-01" and 日期<="2021-12-31"')
tw2021_df
```
