讀寫檔案
open() 回傳一個 file object，而它最常使用的兩個位置引數和一個關鍵字引數是：open(filename, mode, encoding=None)

引入 with...as 的流程控制，在 with block 所執行的 file 物件，最後都會自動執行 f.close 這個動作。

```python
# with...as...可自動 close
# 透過 file實體 可讀取資料
with open("names.txt", encoding='utf-8' ) as file:
    print(file.read())  #傳出字串  #print 螢幕顯示的方式
```

```python
# display 可看到格式
with open("names.txt", encoding='utf-8' ) as file:
    display(file.read())    # display顯示格式 (可看到換行符號)
```

```python
# file.read讀取資料，並儲存到 content，再 key content 來輸出展示
with open("names.txt", encoding='utf-8' ) as file:
    content:str = file.read()

content  #放在最後一行，也有display含意
```

str.split(sep=None, maxsplit=-1)
返回一个由字符串内单词组成的列表，使用 sep 作为分隔字符串。 如果给出了 maxsplit，则最多进行 maxsplit 次拆分（因此，列表最多会有 maxsplit+1 个元素）。 如果 maxsplit 未指定或为 -1，则不限制拆分次数（进行所有可能的拆分）。

```python
with open("names.txt", encoding='utf-8' ) as file:
    content:str = file.read()

content.split(sep='\n')   #用 換行符號 來切割 成 list(可儲存多個值)
```

```python
with open("names.txt", encoding='utf-8' ) as file:
    content:str = file.read()

names:list[str] = content.split(sep='\n')
    
# for...in...迴圈 可放串列資料(list)，明確知道 執行幾次
for name in names[:5]:
    print(name)
```

```python
# 建立空的 list，使用初始化
print(list())
print(list("666"))

# 建立空的 list，使用純值表示法
print([])
print(['0','2','3'])
```

```python
with open("names.txt", encoding='utf-8' ) as file:
    content:str = file.read()

names:list[str] = content.split(sep='\n')
for name in names[:3]:
    print(name)
print("==================")
# subscript語法
print(names[0])
print(names[1])
print(names[2])
print(names[0:2])    #slice (不含最後一個)
print(names[:5])     #slice ->前 5個
print(names[-5:])    #slice ->後 5個
print(names[-1])
```

random.choices(population, weights=None, *, cum_weights=None, k=1)
回傳從 population 中重置取樣出的一個大小為 k 的元素 list。如果 population 為空，則引發 IndexError。

```python
import random
with open('names.txt',encoding='utf-8') as file:
    content:str = file.read()

names:list[str] = content.split(sep='\n')

random.choices(population=names,k=3)
```

```python
import random
import pyinputplus as pyip

with open('names.txt',encoding='utf-8') as file:
    content:str = file.read()

n:int = pyip.inputInt("請輸入取得名字的數量",min=0,max=200)
names:list[str] = content.split(sep='\n')
random_names:list[str] = random.choices(population=names,k=n)
for name in random_names:
    print(name,end=' ')
print()
```
