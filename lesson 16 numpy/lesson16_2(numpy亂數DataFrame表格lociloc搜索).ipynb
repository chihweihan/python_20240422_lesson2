- numpy包在 DataFrame 內，取出numpy
- pandas.DataFrame.values
- Return a Numpy representation of the DataFrame.

```python
import numpy as np
import pandas as pd
array_2d = np.random.randint(50,100,size=(50,5))
df1 = pd.DataFrame(array_2d,
             columns=['國文','英文','數學','自然','社會'],   #欄位
             index = [f'{num}號' for num in range(1,51)],   #索引
             )

df1
```

```python
# 從dataFrame 取出 ndarray
df1.values
```

```python
# dataFrame [欄位名稱]
print(type(df1['國文']))
df1['國文'], df1['英文']

# 可知 dataFrame 由多個 Series 組合起來，Series只有索引(index)，沒有欄位
# -> Series 是一維資料；DataFrame 是二維資料
```

```python
# Series的選取，使用 subscript
# Series [數值的index]
df1['國文'][4]
# Series [數值的slice]
df1['國文'][:5]
# Series[字串的index]
df1['國文']['3號']
# Series[[字串,字串,字串]]
df1['國文'][['1號','2號','5號']]
```

```python
# dataFrame的subscripts
# dataFrame[欄位名稱]
df1['國文']
```

```python
# dataFrame的subscripts：欄的選取
# dataFrame[[欄位名稱,欄位名稱]]
df1[['國文','英文']]
```

### pandas.DataFrame.loc
- 通過標籤或布爾陣列訪問一組行和列。
- loc 用字串；iloc 用索引

```python
# dataFrame的loc：列的選取
# loc[字串]
# iloc [數值]
```

```python
# 選取1號
df1.loc['1號']
```

```python
# 選取多號
# loc[[字串,字串]]
df1.loc[['1號','2號','3號']]
```

```python
# 選取多號
# iloc[index:index]
df1.iloc[:3]
```

```python
# 選取後5位
# loc[數值索引:數值索引]
df1.iloc[-5:]
```
