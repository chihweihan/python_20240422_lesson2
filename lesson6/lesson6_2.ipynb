- 非 連續數值
-兩個條件非相關性

```python
#國文100，同時數學100，獎金1000元
#國文100，或  數學100，獎金500元

#多項選擇，先輸入較嚴格的程式區段

chinese = int(input("請輸入國文分數"))
math = int(input("請輸入數學分數"))

if chinese == 100 and math == 100:
    bonus = 1000
elif chinese == 100 or math == 100:
    bonus = 500
else:
    bonus = 0

print(f"獎金：{bonus}")
```
