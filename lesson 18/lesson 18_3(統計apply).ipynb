```python
import numpy as np
import pandas as pd

score = pd.DataFrame(np.random.randint(50,101,size=(50,5)),
             columns=['國文','英文','數學','自然','社會'],
             index=range(1,51))

score
```

- Apply a function along an axis of the DataFrame.
- 调用一个函数（Python method），让此函数对数据对象进行批量处理。

Lambda函式，也就是匿名函式，不需要定義名稱，只有一行運算式，語法非常簡潔

```python
score.apply(lambda item:print(item))

# 將DataFrame中 所有Series 傳入 apply 並合併成一個 Series
```

```python
def abc(s):
    return s
score.apply(abc)
```

```python
def abc(s):
    return s.max()
score.apply(abc)
```

```python
def abc(s):
    return pd.Series([s.min(),s.max(),s.mean()],index=['最低分','最高分','平均'])
score.apply(abc)
```

```python
# 更換 index 內容
score.index = score.index.map(lambda i:f'{i}號')
score
```

- " ".join()：用指定的符號來連接多個字串

```python
def abc1(item):
    non_pass = item[item < 60]
    non_pass_count = non_pass.count()
    non_str = "/".join(non_pass.index.values.tolist())

    return pd.Series([item.sum(), item.mean(), non_pass_count,non_str], index=['總分','平均','不及格科目數','不及格科目'])
score.apply(abc1, axis=1)   #以橫列做為計算
```
