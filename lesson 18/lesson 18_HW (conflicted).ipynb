```python
#Comprehension 算是 Python 特有的語法，可以用來宣告一些特定 object，或是對這些 object 做一些操作。
#前面的 expression 會變成 key: value 的形式，key 和 value 的部分都可以用 expression 來表達
{key: value for item in iterable (if condition)}
```

```python
import pandas_datareader.data as pdr
import yfinance as yf
import pandas as pd

yf.pdr_override()         #把資料抓到本機
all_stocks = {item:pdr.get_data_yahoo(item) for item in ['2330.TW', '2303.TW', '2454.TW', '2317.TW']}  #抓特定代碼
# 每一個key為一欄
df_stocks = pd.DataFrame({key:df['Adj Close'] for key, df in all_stocks.items()})  #過濾特定data，返回一个视图对象，该对象显示字典的（键，值）元组对的列表。
df_stocks.columns = ["台積電", "聯電", "聯發科","鴻海"]                             #更改column名稱
df_stocks
```
