請收集台積電,聯電,聯發科,鴻海的所有收盤價的資料,成為DataFrame
- 台積電(2330.TW),聯電(2303.TW),聯發科(2454.TW),鴻海(2317.TW)

```python
import pandas_datareader.data as pdr
import yfinance as yf
import pandas as pd
yf.pdr_override()    ##把資料抓到本機

all_stocks = {item:pdr.get_data_yahoo(item) for item in ['2330.TW', '2303.TW', '2454.TW', '2317.TW']}
database = pd.DataFrame({key:df['Adj Close'] for key, df in all_stocks.items()})  #返回一个视图对象，该对象显示字典的（键，值）元组对的列表。
database
```
