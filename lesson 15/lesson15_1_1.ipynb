```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font = FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

student1 = {'':78, '':80, '':92, '':75, '':85} #
student2 = {'':92, '':60, '':85, '':62, '':73}

keys = student1.keys()
subject:list[str] = list(keys)
values = student1.values()
values1 = list(values)
values2 = list(student2.values())  

figure = plt.figure(figsize=(8,5), dpi=250, facecolor='#81C7D4')
axes = figure.add_subplot(1, 1, 1)   #
axes.plot(subject,values1, 'bo--', label="1") # (X,y, '[][marker][style]')  
axes.plot(subject,values2, 'r*-.', label="2")
axes.grid(axis='y')  # (xy)
axes.set_title("1 vs 2",fontproperties=font)
axes.set_xlabel("",fontproperties=font)
axes.set_ylabel("",fontproperties=font)
axes.set_xticklabels(subject,fontproperties=font)
axes.legend([f'1',f'2'],prop=font)   #label

#print(axes.get_xticks())     #
#print(axes.get_yticks())
labels = axes.get_xticklabels()
#print(labels)
for text in labels:
    subject_name = text.get_text()
    score1 = student1[subject_name]
    x_value = text.get_position()[0]
    axes.annotate(str(score1),(x_value, score1+0.5))       #annotate
    score2 = student2[subject_name]
    axes.annotate(str(score2),(x_value, score2+0.5))

plt.show()
```
