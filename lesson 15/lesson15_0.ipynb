繪製圖表                  資料
1. figure實體 -畫布       X軸要呈現的資料
2. Axes實體 -圖表         Y軸要呈現的資料
使用Axes實體修改細節

```python
student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85} #元素數量要記得
student2 = {'chinese':92, 'english':60, 'math':85, 'history':62, 'discover':73}
# dict的實體 ->一定有一個class

keys = student1.keys()
# keys()：返回由字典键组成的一个新视图。
print(type(keys))
print(keys)   #類似list
subject:list[str] = list(keys)
print(subject)
```

```python
values = student1.values()
print(values)  #list，裡面是整數

values1 = list(values)
values2 = list(student2.values())  #實體方法

subject, values1, values2  #傳出tuple
```

```python
# 繪圖  dpi(解析度) facecolor(參考日本色票)
import matplotlib.pyplot as plt
figure = plt.figure(figsize=[8,5], dpi=250, facecolor='#81C7D4')
axes = figure.add_subplot(1, 1, 1)   #一欄一列

axes.plot(subject,values1) # (X軸,y軸)  'bo--'  'ro-.'
axes.plot(subject,values2)

plt.show()
```
