```python
import numpy as np
import pandas as pd
array_2d = np.random.randint(50,100,size=(50,5))
df1 = pd.DataFrame(array_2d,
             columns=['國文','英文','數學','自然','社會'],   #欄位
             index = [f'{num}號' for num in range(1,51)],   #索引
             )

df1
```

```python
# ndarray 最好用的即為 陣列運算
total_se = df1['國文'] + df1['英文'] + df1['數學'] + df1['自然'] + df1['社會']
total_se
```

```python
average_se = total_se / 5
average_se
```

```python
df1['總分'] = total_se
df1['平均'] = average_se
df1
```

```python
# Series 使用比較運算子
df1['總分'] >= 400
```

```python
# 使用 dataFrame/subscript 搜尋資料
# dataFrame[bool Series]

df1[df1['總分'] >= 400]
```

```python
# 查詢平均95分以上
df1[df1['平均'] >= 85]
```

```python
# 查詢平均85分以上 同時數學90分以上
# and = &
# or = |
df1[(df1['平均'] >= 85) & (df1['英文'] >=90)]
```
