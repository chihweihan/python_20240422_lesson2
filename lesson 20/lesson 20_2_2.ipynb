```python
import pandas as pd
df1 = pd.read_csv('每日各站進出站人數2022.csv')
df1.columns = ['日期','code','進站','出站']
df1['日期'] = pd.to_datetime(df1['日期'].astype(str))
#df2 = df1.set_index(keys=['日期','code'])
#df2.columns.name = '進出'
df1
```

class datetime.datetime(year, month, day, hour=0, minute=0, second=0, microsecond=0, tzinfo=None, *, fold=0)
- year, month 和 day 参数是必须的。 tzinfo 可以是 None 或者是一个 tzinfo 子类的实例。 其余的参数必须是在下面范围内的整数：

```python
df1['月份'] = df1['日期'].dt.month
df1
```

```python
df2 = df1.set_index(keys=['日期','code','月份'])
df2.columns.name = '進出'
df2
```

```python
stack1 = df2.stack()
stack1
```

```python
month1 = stack1.groupby(level=['月份','code','進出']).sum()
month1
```

```python
#2022，全台 每月份 各站點 進出站的人口數
month1.unstack('進出')
```
