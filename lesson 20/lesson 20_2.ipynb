```python
import pandas as pd
df1 = pd.read_csv('每日各站進出站人數2022.csv')
df1.columns = ['日期','code','進站','出站']
df1['日期'] = pd.to_datetime(df1['日期'].astype(str)) #將日期格式轉為datetime
df1
```

DataFrame.set_index(keys, *, drop=True, append=False, inplace=False, verify_integrity=False)[source]
- Set the DataFrame index using existing columns.

```python
df2 = df1.set_index(keys=['日期','code'])
df2.columns.name = '進出'
df2  #三維資料
```

```python
df2.info() #索引內是數值 不行
```

```python
stack_data = df2.stack()  #堆疊化
stack_data

# 進站 出站 有重複，可做groupby EX:進站加總
```

```python
#2022年，全台進出站的總人數
stack_data.groupby(level='進出').sum()
```

```python
#2022年，全台各站點 進出站總人口數
sum1 = stack_data.groupby(level=['code','進出']).sum()
sum1.unstack()
```
