json.loads(s, *, cls=None, object_hook=None, parse_float=None, parse_int=None, parse_constant=None, object_pairs_hook=None, **kw)
- 使用这个 转换表 将 s (一个包含 JSON 文档的 str, bytes 或 bytearray 实例) 反序列化为 Python 对象。

```python
import json
import pandas as pd
with open('每日各站進出站人數/車站基本資料集.json',encoding='utf-8') as file:
    content:list[dict] = json.load(file)   #轉成python資料結構
info_df= pd.DataFrame(content)
info_df
```

```python
#篩選資料
info_df1 = info_df.reindex(columns=['stationCode','stationName','stationAddrTw'])
info_df1.columns = ['Code','站名','地址']
info_df1['Code'] = info_df1['Code'].astype(int)   #Series.astype 轉換type型別
info_df1.info()
```

```python
site2023_df = pd.read_csv('每日各站進出站人數/每日各站進出站人數2023.csv')
site2023_df.columns = ['日期','Code','進站','出站']
site2023_df
```

```python
merge2023 = pd.merge(site2023_df,info_df1)
merge2023
```

```python
merge2023.info()
```
